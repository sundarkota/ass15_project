<h3>17. Build a dataset with every possible (UserID, ProductID) pair (cross join)</h3>

```python
products
```

```python
def df_crossjoin(df1, df2, **kwargs):
    df1['_tmpkey'] = 1
    df2['_tmpkey'] = 1

    res = pd.merge(df1, df2, on='_tmpkey', **kwargs).drop('_tmpkey', axis=1)
   # res.index = pd.MultiIndex.from_product((df1.index, df2.index))

    df1.drop('_tmpkey', axis=1, inplace=True)
    df2.drop('_tmpkey', axis=1, inplace=True)

    return res
```

```python
users_prod=df_crossjoin(users,products,suffixes=('_orig', '_dest'))
#users_prod.reset_index()
users_prod.index = users_prod.index.values
users_prod[["UserID","ProductID"]]
```
